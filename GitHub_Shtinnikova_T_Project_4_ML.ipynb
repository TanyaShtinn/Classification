## Task.

2.6. Analyze the feature representing customer balances for the presence of outliers.

The Tukey method, also known as the Tukey's test or the Tukey's range test, is a statistical technique used to identify and handle outliers in a dataset. The Tukey method is based on the concept of the interquartile range (IQR), which is the difference between the 75th percentile (Q3) and the 25th percentile (Q1) of the data.

The steps involved in the Tukey method for outlier detection are as follows:

- Calculate the 25th percentile (Q1) and the 75th percentile (Q3)
- Calculate the interquartile range (IQR):
IQR = Q3 - Q1

- Define the lower and upper bounds for outlier detection:
Lower bound = Q1 - 1.5 * IQR
Upper bound = Q3 + 1.5 * IQR

- Identify outliers:

Any data point that falls below the lower bound or above the upper bound is considered an outlier.
Handling outliers:

Depending on the context and the analysis requirements, you can choose to:
Remove the identified outliers from the dataset.
Winsorize the outliers by capping them at the lower or upper bound.
Retain the outliers and incorporate them into the analysis with appropriate techniques.

```python
# Calculate the 25th percentile (Q1) and the 75th percentile (Q3) 
# of the 'balance' column: 
q1 = df['balance'].quantile(0.25)
q3 = df['balance'].quantile(0.75)

# Calculate the interquartile range (IQR):
iqr = q3 - q1

# Calculate the lower and upper bounds for outlier detection using the Tukey's method:
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
```

```python
lower_bound
```

```python
upper_bound
```

```python
# Identify the values of the balance that go beyond these outliers
outliers = df[(df['balance'] < lower_bound) | (df['balance'] > upper_bound)]
```

```python
# Remove the outliers:
df = df.drop(outliers.index)
```

6.4. We have already tried to build different models, and now it's time to build an ensemble of different types of models.

We will use stacking by combining the algorithms that we have already used before: 
- decision trees, 
- logistic regression and 
- gradient boosting. 

We will use the logistic regression model as a meta-model.

Leave the parameters the same as they were in your algorithms in previous tasks.

- Evaluate the precision metric on a test data.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score

# Creating models for stacking:
model_decision_tree = DecisionTreeClassifier(max_depth=5, criterion='entropy', random_state=42)
model_logistic_regression = LogisticRegression(max_iter=1000, random_state=42, solver='sag')
model_random_forest = RandomForestClassifier(n_estimators=100, criterion='gini', min_samples_leaf=5, max_depth=10, random_state=42)
model_gradient_boost = GradientBoostingClassifier(learning_rate=0.05, n_estimators=300, min_samples_leaf=5, max_depth=5, random_state=42)

# Creating a stacking model:
estimators = [
    ('decision_tree', model_decision_tree),
    ('logistic_regression', model_logistic_regression),
    ('random_forest', model_random_forest),
    ('gradient_boost', model_gradient_boost)
]
meta_model = LogisticRegression(max_iter=1000, random_state=42, solver='sag')
model_stacking = StackingClassifier(estimators=estimators, final_estimator=meta_model)

# Train the stacking model:
model_stacking.fit(X_train_scaled, y_train)

# Calculating accuracy and F1-score on a test data:

y_pred = model_stacking.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)

print(f"Test precision: {precision:.3f}")
print(f"Test accuracy: {accuracy:.3f}")
print(f"Test F1-score: {f1:.3f}")
```

6.6. Previously, we had already considered hyperparameter optimization using GridSearch. However, you know that this is not the only way. One of the more advanced options for hyperparameter optimization is the Optuna framework. Let's apply it to optimize the hyperparameters. For the search, we'll take a random forest and the following parameters:

- n_estimators = trial.suggest_int('n_estimators', 100, 200, 1);
- max_depth = trial.suggest_int('max_depth', 10, 30, 1);
- min_samples_leaf = trial.suggest_int('min_samples_leaf', 2, 10, 1).


Evaluate F1-score on the test data.

```python
# !pip install optuna
```

```python
import optuna
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, accuracy_score

def objective(trial):
    n_estimators = trial.suggest_int('n_estimators', 100, 200, 1)
    max_depth = trial.suggest_int('max_depth', 10, 30, 1)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 2, 10, 1)

    model = RandomForestClassifier(n_estimators=n_estimators,
                                  max_depth=max_depth,
                                  min_samples_leaf=min_samples_leaf,
                                  random_state=42)

    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)

    f1 = f1_score(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    trial.set_user_attr("accuracy", accuracy)

    return f1
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)

print('Best params: ', study.best_params)
print('Best F1-score: ', study.best_value)
print('Best accuracy: ', study.best_trial.user_attrs['accuracy'])

# Best params:  {'n_estimators': 124, 'max_depth': 10, 'min_samples_leaf': 2}
# Best F1-score:  0.8222996515679443
# Best accuracy:  0.831784107946027
```
